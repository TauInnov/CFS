

``` python
def binary_search(L, x):

    while len(L) > 0:
        mid = (len(L) - 1) // 2
        if x == L[mid]:
            return True
        else:
            if x < L[mid]:  # go to left half
                L = L[:mid]
            else:  # go to right half
                L = L[mid+1:]
    return False  # if we got here the search failed
```

### יעילות חיפוש בינארי

נבחן כעת את היעילות של האלגוריתם לחיפוש בינארי, כלומר את מספר האיטרציות שהוא מבצע במקרה הגרוע ובמקרה הטוב.

על כמה איברים מהרשימה עובר האלגוריתם לחיפוש בינארי? נתחיל מהמקרה הטוב, שהוא גם קל מאוד לניתוח.

````python
import json
from jupyterquiz import display_quiz

example = \
'''
[{
    "question": "מהו מספר האיטרציות של חיפוש בינארי במקרה הטוב?",
    "type": "numeric",
    "answers": [
      {
        "type": "value",
        "value": 1,
        "correct": true,
        "feedback": "נכון!"
      },
      {
        "type": "default",
        "feedback": "לא נכון. חשבו - באיזה מקרה נסיים הכי מהר?"
      }
    ]
}]
'''
myquiz = json.loads(example)
display_quiz(myquiz)
````

המקרה הגרוע הוא המעניין יותר, אבל הוא גם מעט מורכב יותר לניתוח.

**בניגוד לחיפוש הסדרתי, הפעם אנחנו לא עוברים על כל איברי הרשימה**, גם לא במקרה הגרוע ביותר. נחשב את מספר האיברים שהאלגוריתם בודק במקרה הגרוע. בכל שלב האלגוריתם בודק איבר אחד, זה שנמצא באמצע טווח החיפוש. אז השאלה היא כמה שלבים מבצע האלגוריתם?

**בחיפוש הבינארי, בכל שלב פוסלים בערך מחצית מהאיברים בטווח החיפוש.** כלומר לרשימה באורך n, אחרי שלב אחד ניוותר עם קצת פחות מ-n/2 איברים. לאחר שני שלבים עם קצת פחות מ-n/4 , לאחר עוד שלב עם קצת פחות מ-n/8 איברים וכך הלאה. כלומר צריך בעצם לשאול - "כמה פעמים אפשר לחצות את המספר n (אורך הרשימה) בטרם נקבל מספר שקטן מ-1?". מספר שקטן מ-1 מייצג את המצב בו בטווח החיפוש יש פחות מאיבר אחד, כלומר יש בו 0 איברים.

במתמטיקה יש סימון לערך מאוד קרוב לתשובה לשאלה שלנו: **log2(n)**, שהוא **מספר הפעמים שצריך לחלק את n ב-2 כדי לקבל בדיוק 1**.   
בחיפוש בינארי צריך להמשיך עוד שלב ולקבל מספר שקטן מ-1 (כי רק אז טווח החיפוש מתרוקן). אבל ההבדל הזה זניח ביחס, להבדל ביעילות בין חיפוש סדרתי לבינארי, במיוחד כאשר מדובר ברשימות ארוכות.  

**לסיכום, בחיפוש בינארי היעילות היא בסדר גודל של log2(n) מכיוון שבסוף כל איטרציה אנו נותרים עם רשימה שאורכה היא כמחצית מהרשימה איתה התחלנו את האיטרציה.  
לעומת זאת, בחיפוש סדרתי היעילות היא בסדר גודל של n מכיוון שכל איטרציה אנו "חותכים" את אורך הרשימה ב1.**  
לדוגמא, נסתכל על המספר n=128. מספר הפעמים שצריך לחלק את 128 ב-2 כדי לקבל 1 הוא בדיוק 7: 128←64←32←16←8←4←2←1   
ואכן, log2(128)=7. לכן, בחיפוש בינארי עבור רשימת קלט בגודל n=128, דרושות 7 איטרציות **לכל היותר** עד שמגיעים לרשימה בגודל 1. לעומת זאת, בחיפוש סדרתי נדרש ל-128 איטרציות במקרה הגרוע ביותר.

````python
import json
from jupyterquiz import display_quiz

example = \
'''
[{
    "question": "מה מספר השלבים הנדרש מהחיפוש הבינארי במקרה הגרוע עבור רשימה בגודל 256?",
    "type": "numeric",
    "answers": [
      {
        "type": "value",
        "value": 9,
        "correct": true,
        "feedback": "נכון! כאשר מגדילים את רשימת הקלט פי 2, חיפוש בינארי מבצע רק איטרציה אחת נוספת בסך הכל!"
      },
      {
        "type": "default",
        "feedback": "לא נכון"
      }
    ]
}]
'''
myquiz = json.loads(example)
display_quiz(myquiz)
````

```{admonition} **כדאי לדעת** 
:class: info
על אלגוריתמים בעלי יעילות **log2(n)** (כמו חיפוש בינארי) נהוג לומר שהם בעלי  **סיבוכיות זמן לוגריתמית**.   
על אלגוריתמים בעלי יעילות **n** (כמו חיפוש בינארי) נהוג לומר שהם בעלי  **סיבוכיות זמן ליניארית**.
```
