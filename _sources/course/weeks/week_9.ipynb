# שבוע 9 — NumPy מתקדם: מטריצות, ערכים עצמיים ויישומים מדעיים

בשבוע התשיעי נעמיק את היכרותנו עם **NumPy**, ספריית היסוד לעבודה מדעית בפייתון.  
לאחר שלמדנו את הבסיס לעבודה עם מערכים, נעבור כעת לשימושים מתקדמים במטריצות,  
פתרון מערכות משוואות ליניאריות, חישוב ערכים ווקטורים עצמיים, ופירוקים מתמטיים חשובים  
הנמצאים בלב תחומים כמו פיזיקה חישובית, עיבוד תמונה, ובינה מלאכותית.

נלמד כיצד להשתמש בכלים החזקים של NumPy לליניאר אלגברה, ונראה דוגמאות חיות:  
מהדמיית דחיסת תמונה באמצעות פירוק לערכים סינגולריים (SVD),  
ועד הערכה הסתברותית של π בעזרת סימולציה מונטה־קרלו.

בסוף השבוע תדעו:
- לבצע חישובים מטריציוניים מתקדמים עם NumPy  
- לפתור מערכות משוואות ליניאריות באופן נומרי  
- לחשב ערכים ווקטורים עצמיים ולהבין את משמעותם הפיזיקלית  
- לבצע פירוק לערכים סינגולריים (SVD) וליישם אותו לבעיות שונות  
- להשתמש בדוגמאות פיזיקליות ומדעיות ליישום מושגים אלו
