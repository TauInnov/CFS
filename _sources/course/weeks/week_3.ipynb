# שבוע 3 — רשימות

בשבוע השלישי נעבור מעבודה עם משתנים בודדים לעבודה עם **אוספים של נתונים** — ובראשם *רשימות (Lists)*,  
אחת מהמבנים המרכזיים והחזקים ביותר בפייתון. רשימות מאפשרות לנו לשמור סדרות של פריטים, לשנותן בזמן ריצה,  
ולבצע עליהן פעולות בקלות וביעילות.

נלמד כיצד ליצור רשימות, לגשת לאיברים באמצעות אינדקסים, להשתמש באופרטורים ייעודיים,  
ולהכיר פונקציות מובנות המקלות על עבודה עם רשימות.  
נעמיק בשיטות (methods) של אובייקטים מסוג list, נראה כיצד ניתן לקנן רשימות זו בתוך זו,  
ונסיים בסדרת דוגמאות מעשיות המדגימות כיצד רשימות משתלבות בלוגיקה של תוכניות אמיתיות.

בסוף השבוע תדעו:
- ליצור ולעדכן רשימות ולהשתמש באינדקסים כדי לגשת לאיברים  
- לבצע פעולות כמו חיבור, חיתוך ומיון של רשימות  
- להשתמש בפונקציות ובשיטות מובנות לעיבוד נתונים  
- לעבוד עם רשימות מקוננות (רשימות בתוך רשימות)  
- לכתוב קוד יעיל וקצר יותר בזכות מבני נתונים דינמיים
