def circle_area(radius):
    """Compute the area of a circle with non-negative radius."""
    if radius < 0:
        raise ValueError("Radius must be non-negative")
    return math.pi * radius ** 2


class TestCircleArea(unittest.TestCase):
    def setUp(self):

        self.test_cases = [

    def test_circle_area_values(self):
        """Check multiple radius values using subTest."""
        for radius, expected in self.test_cases:
            with self.subTest(radius=radius):
                result = circle_area(radius)
                self.assertAlmostEqual(result, expected, places=6)

    def test_negative_radius_raises(self):
        """Verify that negative radius raises ValueError."""
        with self.assertRaises(ValueError):
            circle_area(-1)
