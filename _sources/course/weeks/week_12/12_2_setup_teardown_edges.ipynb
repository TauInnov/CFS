## דפוס Arrange–Act–Assert (AAA)

זהו סגנון כתיבה שמסייע לשמור על קריאות הבדיקה:

1. **Arrange** – הגדרת הנתונים והאובייקטים.  
2. **Act** – הפעלת הפעולה שאנו בודקים.  
3. **Assert** – בדיקה שהתוצאה תואמת לציפיות.

### דוגמה:

שלושת השלבים הקלאסיים של כל בדיקה:

1. **Arrange (הכנה)**  
   בשלב זה אנחנו מגדירים את נתוני הקלט ואת התוצאה הצפויה — במקרה הזה, בסיס (`base`) וגובה (`height`) של משולש,  
   ואת הערך הצפוי של שטחו לפי הנוסחה \( A = \frac{1}{2} \times base \times height \).

3. **Assert (בדיקה)**
